```python
import csv
from PIL import Image
import numpy as np
import string
import os

file_path = "/Users/ishantkamboj/Documents/Chandigarh university/SEM_5/AML/Exp/Exp_7_new/archive(4)/A_Z Handwritten Data.csv"

count = 1
last_digit = None

image_folder = "/Users/ishantkamboj/Documents/Chandigarh university/SEM_5/AML/Exp/Experiment_7/Converted_images"

Alphabet_mapping_list = list(string.ascii_uppercase)

for alphabet in Alphabet_mapping_list:
    path = image_folder + '/' + alphabet
    if not os.path.exists(path):
        os.makedirs(path)

with open(file_path, newline='') as csvfile:
        reader = csv.reader(csvfile, delimiter=',', quotechar='|')
        count = 0
        for row in reader:
            digit_name = row.pop(0)
            image_array = np.asarray(row)
            image_array = image_array.reshape(28,28)
            new_image = Image.fromarray(image_array.astype('uint8'))
            
            if last_digit != str(Alphabet_mapping_list[(int)(digit_name)]):
                last_digit = str(Alphabet_mapping_list[(int)(digit_name)])
                count = 0
                print("")
                print("processing alphabet - "+ str(last_digit))
            
            image_path = image_folder + '/' + last_digit + '/' + str(last_digit) + '-' + str(count) + '.png'
            new_image.save(image_path)
            count = count+1
            
            if count%1000 == 0:
                print("Image processed: " + str(count))
            
            
            
            
            
```
